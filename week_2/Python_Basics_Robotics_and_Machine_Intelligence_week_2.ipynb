## 2) Variables
A variable stores a value.

Guidelines:
- Use **meaningful names** (e.g., `robot_speed`)
- Python is **case-sensitive** (`speed` ≠ `Speed`)


```python
robot_speed = 1.2
robot_name = "Atlas"
is_autonomous = True

print(robot_name, "speed =", robot_speed)
print("Autonomous?", is_autonomous)
```

## 3) Data Types
Common types:
- `int`, `float`, `str`, `bool`

Use `type()` to check.


```python
x = 10
y = 2.5
z = "robot"
w = False

print(type(x), type(y), type(z), type(w))
```

## 4) Operators
### Arithmetic
- `+  -  *  /  //  %  **`

Example: distance = speed × time


```python
speed = 1.5
time_sec = 8
distance = speed * time_sec
print("Distance:", distance, "m")

print("10 / 3 =", 10 / 3)
print("10 // 3 =", 10 // 3)
print("10 % 3 =", 10 % 3)
print("2 ** 5 =", 2 ** 5)
```

## 5) Strings
Useful operations:
- concatenation: `+`
- repetition: `*`
- length: `len()`
- indexing: `s[0]`


```python
robot = "cobot"
print(robot + "s")
print(robot * 3)
print("Length:", len(robot))
print("First letter:", robot[0])
```

### f-strings (recommended)


```python
name = "Pepper"
battery = 87
print(f"Robot {name} battery = {battery}%")
```

## 7) Booleans & Comparisons
Comparisons: `== != < <= > >=`

Logical operators: `and`, `or`, `not`


```python
speed = 1.2
print(speed > 1.0)
print(speed == 1.2)

has_sensor = True
battery_ok = False
print("Can start?", has_sensor and battery_ok)
print("Need charging?", not battery_ok)
```

## 8) Conditionals: `if / elif / else`


```python
battery = 25

if battery >= 80:
    print("Battery high: full operation.")
elif battery >= 30:
    print("Battery medium: limit speed.")
else:
    print("Battery low: return to charging station.")
```

## 9) Loops
### `for` loop


```python
for i in range(5):
    print("Step", i)
```

### `while` loop


```python
battery = 5
while battery > 0:
    print("Battery:", battery)
    battery -= 1
print("Shutdown")
```

## 10) Lists
Lists are mutable collections.


```python
sensors = ["camera", "lidar", "imu"]
print(sensors)
print("First:", sensors[0])

sensors.append("ultrasonic")
print("Updated:", sensors)
print("Count:", len(sensors))
```

### Slicing
`list[start:stop]` (stop not included)


```python
values = [10, 20, 30, 40, 50]
print(values[1:4])
print(values[:3])
print(values[3:])
```

## 11) Tuples
Tuples are immutable.


```python
position = (2.0, 1.5)
print(position)
print("x =", position[0])
```

## 12) Dictionaries
Key → value pairs.


```python
robot = {
    "name": "R2",
    "speed": 1.1,
    "battery": 72,
    "sensors": ["camera", "imu"]
}

print(robot["name"])
robot["battery"] = 68
print(robot)
```

## 13) Sets
Unique items only.


```python
sensor_types = {"camera", "lidar", "camera"}
print(sensor_types)
sensor_types.add("imu")
print(sensor_types)
```

## 16) Solutions

```python
speed = 1.6
time_s = 12
print("A) distance =", speed * time_s, "m")

distance_to_human = 0.4
if distance_to_human < 0.5:
    print("B) TOO CLOSE")
else:
    print("B) SAFE")

sensors = ["camera", "lidar", "imu"]
for idx, s in enumerate(sensors):
    print(f"C) {idx}: {s}")

battery = 3
while battery > 0:
    print("D) battery", battery)
    battery -= 1
print("D) shutdown")
```
